##Recorrer cada Estructura con un ciclo FOR

```python
# lista
lista_frutas = ["manzana", "banana", "cereza", "naranja", "pera"]  # lista
lista_animales = ["gato", "perro", "elefante", "tigre", "leon"]  # lista

print("Frutas:")
for fruta in lista_frutas:
    print(fruta)

print("\nAnimales:")
for animal in lista_animales:
    print(animal)

print("\n")

# tupla
tupla_verduras = ("zanahoria", "cebolla", "tomate", "lechuga", "pimenton")  # tupla
tupla_colores = ("rojo", "azul", "verde", "amarillo", "morado")  # tupla

print("Verduras:")
for verdura in tupla_verduras:
    print("Me gusta la", verdura)

print("\nColores:")
for color in tupla_colores:
    print("Mi color preferido es", color)

print("\n")

# conjunto
conjunto_electrodomesticos = {"Televisor", "Computador", "Nevera", "Lavadora", "Cafetera"}  # conjunto
conjunto_paises = {"España", "Francia", "Alemania", "Italia", "Reino Unido"}  # conjunto

print("Electrodomésticos:")
for electro in conjunto_electrodomesticos:
    print(electro)

print("\nPaises:")
for pais in conjunto_paises:
    print(pais)
```
