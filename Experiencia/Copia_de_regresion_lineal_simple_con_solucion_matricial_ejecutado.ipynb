```python
import pandas as pd
x = [1,2,2,3,4,4,5,6]
y = [2,3,4,4,4,6,5,7]

d = {"x":x, "y":y}
df = pd.DataFrame(d)
df
```

```python
X=np.array(df2)
X.shape
```

```python
Xpinv= np.linalg.pinv(X)
Xpinv.shape
```

```python
y = np.array(df[["y"]])
y.shape
```

```python
betas = Xpinv @ y
print(betas.shape)
betas
```

# Cálculo de Errores (manualmente)

```python
df["ypred"] = X @ betas
df
```

```python
# error = y - ypred
df["y-ypred"] =  df["y"] - df["ypred"]
df
```

```python
# error = y - ypred
df["error^2"] = df["y-ypred"]**2
df
```

```python
error_cuad_total = df["error^2"].sum()
error_cuad_total
```

```python
# MSE = error_cuad_medio
mse = error_cuad_total / len(df)
mse
```

```python
rmse = mse**(1/2)
rmse
```

```python
# Calculate predicted values
df["ypred"] = X @ betas

# Calculate error
df["y-ypred"] =  df["y"] - df["ypred"]

# Calculate absolute error
df["abs_error"] = abs(df["y-ypred"])
display(df)

# Calculate total absolute error
total_abs_error = df["abs_error"].sum()
print(f"Total Absolute Error: {total_abs_error}")

# Calculate Mean Absolute Error (MAE)
mae = total_abs_error / len(df)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate Absolute Percentage Error
df["abs_perc_error"] = (df["abs_error"] / df["y"]) * 100
display(df)

# Calculate Mean Absolute Percentage Error (MAPE)
mape = df["abs_perc_error"].mean()
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")

# Calculate squared error
df["error^2"] = df["y-ypred"]**2

# Calculate R2
# Calculate the mean of y
y_mean = df["y"].mean()

# Calculate the total sum of squares (SST)
df["y-y_mean"] = df["y"] - y_mean
df["(y-y_mean)^2"] = df["y-y_mean"]**2
sst = df["(y-y_mean)^2"].sum()

# The sum of squared errors (SSE) is already calculated as error_cuad_total
sse = df["error^2"].sum()

# Calculate R2
r2 = 1 - (sse / sst)
print(f"R2 Score: {r2}")
display(df)
```

## Análisis de Métricas de Error y Rendimiento del Modelo

En las celdas anteriores, calculamos varias métricas para evaluar el rendimiento de nuestro modelo de regresión. A continuación, se presenta un análisis de cada una de ellas en español:

**Error Cuadrático Total (Sum of Squared Errors - SSE):**

- **Concepto:** Es la suma de los cuadrados de las diferencias entre los valores reales de 'y' y los valores predichos por el modelo ('ypred').
- **Significado:** Representa la variabilidad total de los errores del modelo. Un valor menor indica que las predicciones del modelo están más cerca de los valores reales.
- **Representación en las cifras:** En este caso, el `error_cuad_total` es `3.5471698113207553`.

**Error Cuadrático Medio (Mean Squared Error - MSE):**

- **Concepto:** Es el promedio de los errores cuadráticos. Se calcula dividiendo el Error Cuadrático Total por el número de observaciones.
- **Significado:** Mide la dispersión promedio de los errores al cuadrado. Al igual que el SSE, un valor menor indica un mejor ajuste del modelo.
- **Representación en las cifras:** El `mse` es `0.4433962264150944`.

**Raíz del Error Cuadrático Medio (Root Mean Squared Error - RMSE):**

- **Concepto:** Es la raíz cuadrada del MSE.
- **Significado:** Proporciona una medida del error en las mismas unidades que la variable de respuesta ('y'). Es más interpretable que el MSE porque no está en unidades cuadradas. Un RMSE más bajo indica un mejor rendimiento del modelo.
- **Representación en las cifras:** El `rmse` es `0.6658800390574074`.

**Error Absoluto Total:**

- **Concepto:** Es la suma de los valores absolutos de las diferencias entre los valores reales de 'y' y los valores predichos por el modelo ('ypred').
- **Significado:** Representa la magnitud total de los errores sin considerar su dirección.
- **Representación en las cifras:** El `total_abs_error` es `4.56603773584906`.

**Error Absoluto Medio (Mean Absolute Error - MAE):**

- **Concepto:** Es el promedio de los errores absolutos. Se calcula dividiendo el Error Absoluto Total por el número de observaciones.
- **Significado:** Mide la magnitud promedio de los errores sin considerar su dirección. Es menos sensible a valores atípicos que el MSE o RMSE. Un MAE más bajo indica un mejor rendimiento del modelo.
- **Representación en las cifras:** El `mae` es `0.5707547169811324`.

**Error Porcentual Absoluto Medio (Mean Absolute Percentage Error - MAPE):**

- **Concepto:** Es el promedio de los errores porcentuales absolutos. Se calcula como la suma de los errores absolutos dividida por los valores reales de 'y', y luego multiplicada por 100.
- **Significado:** Mide el error promedio como un porcentaje de los valores reales. Es útil para comparar el rendimiento del modelo en diferentes conjuntos de datos o variables con diferentes escalas. Un MAPE más bajo indica un mejor rendimiento. Es importante tener precaución con este métrica si hay valores de 'y' cercanos a cero.
- **Representación en las cifras:** El `mape` es `13.463050314465423%`.

**R-cuadrado (R2 Score):**

- **Concepto:** También conocido como coeficiente de determinación. Mide la proporción de la varianza en la variable dependiente ('y') que es predecible a partir de la(s) variable(s) independiente(s) ('x') en un modelo de regresión.
- **Significado:** Indica qué tan bien se ajusta el modelo a los datos. Un valor de R2 de 1.0 significa que el modelo explica toda la variabilidad en 'y', mientras que un valor de 0.0 significa que el modelo no explica ninguna variabilidad. Un R2 más alto generalmente indica un mejor ajuste.
- **Representación en las cifras:** El `r2` es `0.8015569336324053`. Esto significa que aproximadamente el 80.16% de la variabilidad en 'y' es explicada por el modelo.

En resumen, estas métricas nos dan una idea cuantitativa de qué tan bien nuestro modelo de regresión lineal se ajusta a los datos. Un RMSE y MAE bajos, junto con un R2 alto, sugieren un buen rendimiento del modelo en este conjunto de datos.
