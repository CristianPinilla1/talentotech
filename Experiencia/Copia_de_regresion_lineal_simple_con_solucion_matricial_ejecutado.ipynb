```python
import pandas as pd
x = [1,2,2,3,4,4,5,6]
y = [2,3,4,4,4,6,5,7]

d = {"x":x, "y":y}
df = pd.DataFrame(d)
df
```

```python
X=np.array(df2)
X.shape
```

```python
Xpinv= np.linalg.pinv(X)
Xpinv.shape
```

```python
y = np.array(df[["y"]])
y.shape
```

```python
betas = Xpinv @ y
print(betas.shape)
betas
```

# Clculo de Errores (manualmente)

```python
df["ypred"] = X @ betas
df
```

```python
# error = y - ypred
df["y-ypred"] =  df["y"] - df["ypred"]
df
```

```python
# error = y - ypred
df["error^2"] = df["y-ypred"]**2
df
```

```python
error_cuad_total = df["error^2"].sum()
error_cuad_total
```

```python
# MSE = error_cuad_medio
mse = error_cuad_total / len(df)
mse
```

```python
rmse = mse**(1/2)
rmse
```

```python
# Calculate predicted values
df["ypred"] = X @ betas

# Calculate error
df["y-ypred"] =  df["y"] - df["ypred"]

# Calculate absolute error
df["abs_error"] = abs(df["y-ypred"])
display(df)

# Calculate total absolute error
total_abs_error = df["abs_error"].sum()
print(f"Total Absolute Error: {total_abs_error}")

# Calculate Mean Absolute Error (MAE)
mae = total_abs_error / len(df)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate Absolute Percentage Error
df["abs_perc_error"] = (df["abs_error"] / df["y"]) * 100
display(df)

# Calculate Mean Absolute Percentage Error (MAPE)
mape = df["abs_perc_error"].mean()
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")

# Calculate squared error
df["error^2"] = df["y-ypred"]**2

# Calculate R2
# Calculate the mean of y
y_mean = df["y"].mean()

# Calculate the total sum of squares (SST)
df["y-y_mean"] = df["y"] - y_mean
df["(y-y_mean)^2"] = df["y-y_mean"]**2
sst = df["(y-y_mean)^2"].sum()

# The sum of squared errors (SSE) is already calculated as error_cuad_total
sse = df["error^2"].sum()

# Calculate R2
r2 = 1 - (sse / sst)
print(f"R2 Score: {r2}")
display(df)
```

## Anlisis de Mtricas de Error y Rendimiento del Modelo

En las celdas anteriores, calculamos varias mtricas para evaluar el rendimiento de nuestro modelo de regresin. A continuacin, se presenta un anlisis de cada una de ellas en espaol:

**Error Cuadrtico Total (Sum of Squared Errors - SSE):**

- **Concepto:** Es la suma de los cuadrados de las diferencias entre los valores reales de 'y' y los valores predichos por el modelo ('ypred').
- **Significado:** Representa la variabilidad total de los errores del modelo. Un valor menor indica que las predicciones del modelo estn ms cerca de los valores reales.
- **Representacin en las cifras:** En este caso, el `error_cuad_total` es `3.5471698113207553`.

**Error Cuadrtico Medio (Mean Squared Error - MSE):**

- **Concepto:** Es el promedio de los errores cuadrticos. Se calcula dividiendo el Error Cuadrtico Total por el nmero de observaciones.
- **Significado:** Mide la dispersin promedio de los errores al cuadrado. Al igual que el SSE, un valor menor indica un mejor ajuste del modelo.
- **Representacin en las cifras:** El `mse` es `0.4433962264150944`.

**Raz del Error Cuadrtico Medio (Root Mean Squared Error - RMSE):**

- **Concepto:** Es la raz cuadrada del MSE.
- **Significado:** Proporciona una medida del error en las mismas unidades que la variable de respuesta ('y'). Es ms interpretable que el MSE porque no est en unidades cuadradas. Un RMSE ms bajo indica un mejor rendimiento del modelo.
- **Representacin en las cifras:** El `rmse` es `0.6658800390574074`.

**Error Absoluto Total:**

- **Concepto:** Es la suma de los valores absolutos de las diferencias entre los valores reales de 'y' y los valores predichos por el modelo ('ypred').
- **Significado:** Representa la magnitud total de los errores sin considerar su direccin.
- **Representacin en las cifras:** El `total_abs_error` es `4.56603773584906`.

**Error Absoluto Medio (Mean Absolute Error - MAE):**

- **Concepto:** Es el promedio de los errores absolutos. Se calcula dividiendo el Error Absoluto Total por el nmero de observaciones.
- **Significado:** Mide la magnitud promedio de los errores sin considerar su direccin. Es menos sensible a valores atpicos que el MSE o RMSE. Un MAE ms bajo indica un mejor rendimiento del modelo.
- **Representacin en las cifras:** El `mae` es `0.5707547169811324`.

**Error Porcentual Absoluto Medio (Mean Absolute Percentage Error - MAPE):**

- **Concepto:** Es el promedio de los errores porcentuales absolutos. Se calcula como la suma de los errores absolutos dividida por los valores reales de 'y', y luego multiplicada por 100.
- **Significado:** Mide el error promedio como un porcentaje de los valores reales. Es til para comparar el rendimiento del modelo en diferentes conjuntos de datos o variables con diferentes escalas. Un MAPE ms bajo indica un mejor rendimiento. Es importante tener precaucin con este mtrica si hay valores de 'y' cercanos a cero.
- **Representacin en las cifras:** El `mape` es `13.463050314465423%`.

**R-cuadrado (R2 Score):**

- **Concepto:** Tambin conocido como coeficiente de determinacin. Mide la proporcin de la varianza en la variable dependiente ('y') que es predecible a partir de la(s) variable(s) independiente(s) ('x') en un modelo de regresin.
- **Significado:** Indica qu tan bien se ajusta el modelo a los datos. Un valor de R2 de 1.0 significa que el modelo explica toda la variabilidad en 'y', mientras que un valor de 0.0 significa que el modelo no explica ninguna variabilidad. Un R2 ms alto generalmente indica un mejor ajuste.
- **Representacin en las cifras:** El `r2` es `0.8015569336324053`. Esto significa que aproximadamente el 80.16% de la variabilidad en 'y' es explicada por el modelo.

En resumen, estas mtricas nos dan una idea cuantitativa de qu tan bien nuestro modelo de regresin lineal se ajusta a los datos. Un RMSE y MAE bajos, junto con un R2 alto, sugieren un buen rendimiento del modelo en este conjunto de datos.
