```python

# Importando la Libreria/biblioteca de matem치ticas de Python
import math

# DEFINICION DE LAS FUNCIONES:
def mostrar_t칤tulo():
  print('Ejemplo Calculadora Sencilla (+,-,*,/) por Felipe Escall칩n:')

def mostrar_menu():
  print()
  print("MENU:")
  print("1: Sumar (+)")
  print("2: Restar (-)")
  print("3: Multiplicar (*)")
  print("4: Dividir (/)")
  print("5: Logaritmo (log10)")
  print("6: Coseno (cos)")
  print("7: Tangente (tan)")
  print("8: Seno (sen)")
  print("9: Potenciacion (x^2)")
  print("10: Raiz cuadrada")
  print("11: Salir")

def ingresar_datos():
  opcion = input("Ingrese la opci칩n deseada del men칰: ")
  if opcion == "0": # Para salir
    return opcion, None, None
  #Validar que la opcion exista en el men칰
  if opcion <"1" or opcion >"11":
    return opcion, None, None
  #Si es valida, se piden los n칰meros
  numero_uno = float(input("Ingrese el primer n칰mero de la operaci칩n: "))
  numero_dos = float(input("Ingrese el segundo n칰mero de la operaci칩n: "))
  return opcion, numero_uno, numero_dos

# FUNCIONES DE OPERACIONES
def calcular_suma(numero_uno, numero_dos): return numero_uno + numero_dos
def calcular_resta(numero_uno, numero_dos): return numero_uno - numero_dos
def calcular_multiplicacion(numero_uno, numero_dos): return numero_uno * numero_dos
def calcular_division(numero_uno, numero_dos): return numero_uno / numero_dos
def calcular_logaritmo_base10(numero): return math.log10(numero)
def calcular_coseno(angulo): return math.cos(angulo)
def calcular_tangente(angulo): return math.tan(angulo)
def calcular_seno(angulo): return math.sin(angulo)
def calcular_potenciacion(numero): return numero**2
def calcular_raiz_cuadrada(numero): return math.sqrt(numero)

# USANDO MATCH-CASE
def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos):
  match opcion:
    case "1" | "+" | "Sumar":
      print("Vamos a sumar (+):")
      print("Resultado:", calcular_suma(numero_uno, numero_dos))
    case "2" | "-" | "Restar":
      print("Vamos a restar (-):")
      print("Resultado:", calcular_resta(numero_uno, numero_dos))
    case "3" | "*" | "Multiplicar":
      print("Vamos a multiplicar (*):")
      print("Resultado:", calcular_multiplicacion(numero_uno, numero_dos))
    case "4" | "/" | "Dividir":
      print("Vamos a dividir (/):")
      if numero_dos != 0:
        print("Resultado:", calcular_division(numero_uno, numero_dos))
      else:
        print("丘멆잺 춰No es posible dividir entre cero!")
    case "5" | "log10" | "logaritmo":
      print("Logaritmo base 10 de los valores:")
      print("Resultado1:", calcular_logaritmo_base10(numero_uno))
      print("Resultado2:", calcular_logaritmo_base10(numero_dos))
    case "6" | "cos" | "Coseno":
      print("Coseno de los valores:")
      print("Resultado1:", calcular_coseno(numero_uno))
      print("Resultado2:", calcular_coseno(numero_dos))
    case "7" | "tan" | "Tangente":
      print("Tangente de los valores:")
      print("Resultado1:", calcular_tangente(numero_uno))
      print("Resultado2:", calcular_tangente(numero_dos))
    case "8" | "sen" | "Seno":
      print("Seno de los valores:")
      print("Resultado1:", calcular_seno(numero_uno))
      print("Resultado2:", calcular_seno(numero_dos))
    case "9" | "potenciacion" | "potencia":
      print("Potenciaci칩n (x^2) de los valores:")
      print("Resultado1:", calcular_potenciacion(numero_uno))
      print("Resultado2:", calcular_potenciacion(numero_dos))
    case "10" | "raiz" | "raiz cuadrada":
      print("Ra칤z cuadrada de los valores:")
      print("Resultado1:", calcular_raiz_cuadrada(numero_uno))
      print("Resultado2:", calcular_raiz_cuadrada(numero_dos))
    case _:
      print("丘멆잺 Opci칩n inv치lida. Intenta de nuevo.")

def llamar_funciones():
  mostrar_t칤tulo()
  while True: # Ciclo While para el men칰
    mostrar_menu()
    opcion, numero_uno, numero_dos = ingresar_datos()
    if opcion == "11":
      print("游녦 Saliendo de la calculadora. 춰Hasta luego!")
      break
    if numero_uno is None or numero_dos is None:
      print("丘멆잺 Opci칩n inv치lida. Intenta de nuevo.")
      continue
    procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos)

# EJECUCION DEL PROGRAMA:
llamar_funciones()
```
