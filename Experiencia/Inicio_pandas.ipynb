# Task
Create a dataframe named `df` with 100 rows containing names and their respective data. Create a new dataframe named `df_100` from `df` and add a column named 'Clasificación Peso' to `df_100` that classifies individuals based on their weight (overweight, normal weight, underweight).

## Calculate bmi

### Subtask:
Calculate the Body Mass Index (BMI) for each person in the `df_100` DataFrame using their weight and height.


**Reasoning**:
Calculate the BMI using the provided formula and store it in a new column named 'BMI' in the `df_100` DataFrame.



```python
df_100['BMI'] = df_100['Peso'] / (df_100['Altura'] ** 2)
df_100.head()
```

## Define weight classification function

### Subtask:
Create a function to classify individuals based on their BMI.


**Reasoning**:
Define a function to classify BMI values based on standard ranges.



```python
def classify_bmi(bmi):
  """Classifies BMI into categories."""
  if bmi < 18.5:
    return 'bajo peso'
  elif 18.5 <= bmi < 25:
    return 'peso normal'
  elif 25 <= bmi < 30:
    return 'sobrepeso'
  else:
    return 'obesidad'
```

## Apply classification

### Subtask:
Apply the classification function to the `df_100` DataFrame to create the new 'Clasificación Peso' column.


**Reasoning**:
Apply the `classify_bmi` function to the 'BMI' column of the `df_100` DataFrame and store the results in a new column named 'Clasificación Peso'.



```python
df_100['Clasificación Peso'] = df_100['BMI'].apply(classify_bmi)
display(df_100.head())
```

## Display updated dataframe

Display the updated `df_100` DataFrame with the new classification column.


**Reasoning**:

```python
# El método .describe() en columnas categóricas nos da:
# - count: cuántos datos no nulos hay
# - unique: cuántas categorías diferentes existen
# - top: cuál es la categoría más frecuente
# - freq: cuántas veces aparece esa categoría más frecuente
display(df_100[["Nombre", "Clasificación Peso"]].describe())
```

```python
# 2. Frecuencias absolutas con .value_counts()
# ==========================================

# .value_counts() cuenta cuántas veces aparece cada categoría.
# Esto sirve para ver la distribución de una variable categórica.
# Ejemplo: cuántas personas están en cada categoría de IMC.
display(df_100["Clasificación Peso"].value_counts())

# También se puede aplicar a "Nombre" (aunque puede haber muchos nombres).
# Aquí mostramos solo los 10 más frecuentes.
print(df_100["Nombre"].value_counts().head(10))
```

```python
# 3. Frecuencias relativas (porcentajes)

# .value_counts(normalize=True) calcula proporciones (entre 0 y 1).
# Multiplicamos por 100 para convertir a porcentajes.
# Esto sirve para entender qué porcentaje de la muestra está en cada categoría.
display(df_100["Clasificación Peso"].value_counts(normalize=True) * 100)
```

```python
# 4. Tablas cruzadas con pd.crosstab()

# pd.crosstab() permite cruzar dos variables categóricas.
# Aquí, por ejemplo, mostramos cuántas veces aparece cada combinación
# de Nombre y Clasificación. (Normalmente útil si hay nombres repetidos).
display(pd.crosstab(df_100["Nombre"], df_100["Clasificación Peso"]))
```

```python
# 5. Visualización de las categorías

# Con .plot(kind="bar") podemos graficar los conteos de las categorías.
# Esto sirve para visualizar rápidamente la distribución de la variable.
# En este caso, la distribución de personas según su clasificación de IMC.
df_100["Clasificación Peso"].value_counts().plot(kind="bar", title="Distribución de Clasificación IMC")
```
